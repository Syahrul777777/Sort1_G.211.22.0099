Bubble Sort

```python
def SinkingSort(arr): #Mendefinisikan fungsi SinkingSort yang menerima sebuah array sebagai parameter.
    x = len(arr) #Menghitung panjang (jumlah elemen) dari array arr dan menyimpannya dalam variabel x.
    swapped = False #Membuat variabel swapped yang akan digunakan untuk memeriksa apakah ada pertukaran elemen dalam iterasi terakhir.

    for i in range(x-1): #Iterasi melalui array hingga elemen kedua terakhir.
        for j in range(0, x-i-1): #Iterasi untuk membandingkan dan menukar elemen yang tidak terurut.
            if arr[j] > arr[j + 1]: #Memeriksa apakah elemen saat ini lebih besar dari elemen berikutnya.

                swapped = True # Mengatur swapped menjadi True karena terjadi pertukaran.
                arr[j], arr[j + 1] = arr[j + 1], arr[j] #Menukar posisi elemen yang tidak terurut.

        if not swapped: #Jika tidak ada pertukaran dalam satu iterasi, array sudah terurut, dan loop dihentikan.

            break #Menghentikan iterasi jika array sudah terurut

arr = [76, 27, 19, 45, 88, 90, 15] #Mendefinisikan array yang akan diurutkan.
SinkingSort(arr) #Memanggil fungsi SinkingSort untuk mengurutkan array
print("Sorted array using Sinking Sort:")
for i in range(len(arr)):
    print("%d" % arr[i], end=" ") #Mencetak array yang sudah terurut.
```

Selection Sort

```python
import sys #Mengimpor modul sys, tetapi tidak digunakan dalam kode ini.

S = [84, 29, 15, 37, 19] # Mendefinisikan array yang akan diurutkan

for i in range(len(S)): # Iterasi melalui setiap elemen dalam array

    min_idx = i  # Inisialisasi indeks minimum


    for j in range(i + 1, len(S)):  # Iterasi untuk mencari elemen terkecil dalam sisa array
        if S[min_idx] > S[j]:   # Memeriksa apakah elemen saat ini lebih kecil dari elemen terkecil yang ditemukan sejauh ini
            min_idx = j #Memperbarui indeks minimum jika ditemukan elemen yang lebih kecil.

    S[i], S[min_idx] = S[min_idx], S[i]  # Swap elemen terkecil dengan elemen pertama dalam sisa array

# Mencetak array yang sudah terurut
print("Sorted array:") #Mencetak pesan sebelum mencetak array yang sudah terurut.
for i in range(len(S)): #Iterasi melalui setiap elemen dalam array yang sudah terurut.
    print("%d" % S[i], end=" ") #Mencetak setiap elemen dalam array yang sudah terurut dengan format tertentu.
```

Insection Sort


```python
def insertionSort(arr): #Mendefinisikan fungsi Insertion Sort yang menerima sebuah array sebagai parameter.
    for i in range(1, len(arr)): #Iterasi dimulai dari indeks kedua (indeks 1) hingga akhir array.
        key = arr[i] #Mengambil elemen saat ini sebagai kunci yang akan disisipkan ke dalam array yang telah diurutkan.
        s = i - 1 #Inisialisasi indeks s untuk membandingkan dengan elemen-elemen sebelumnya.
        while s >= 0 and key < arr[s]: #Memeriksa dan memindahkan elemen yang lebih besar ke posisi setelahnya.
            arr[s + 1] = arr[s] #Memindahkan elemen yang lebih besar ke posisi setelahnya.
            s -= 1 #Memperbarui indeks s untuk memeriksa elemen sebelumnya.
        arr[s + 1] = key #Menempatkan elemen kunci pada posisi yang tepat dalam array yang telah diurutkan.

# Driver code to test above
arr = [22, 37, 7, 9, 3] #Mendefinisikan array yang akan diurutkan.
insertionSort(arr) #Memanggil fungsi Insertion Sort untuk mengurutkan array.

for i in range(len(arr)): #Iterasi melalui setiap elemen dalam array yang sudah terurut.
    print("%d" % arr[i])  #Mencetak setiap elemen dalam array yang sudah terurut dengan format tertentu.
```
